```python
import pandas as pd
df = pd.read_csv(r"C:\Users\CSE Lab\OneDrive\Desktop\student_data.csv")
print(df)
```

```python
print("Data Types:")
print(df.dtypes)
```

```python
print("Dataset info:")
df.info()
```

```python
print("Task-B\n")
print("Mean Attendance:",df["Attendance"].mean())
print("Average Internal Marks:",df["InternalMarks"].mean())
print("Average External Marks:",df["ExternalMarks"].mean())
print("Maximum Internal Marks:",df["InternalMarks"].max())
print("Maximum External Marks:",df["ExternalMarks"].max())
print("Minimum Internal Marks:",df["InternalMarks"].min())
print("Minimum External Marks:",df["ExternalMarks"].min())
print("Standard Deviation of Internal Marks:",df["InternalMarks"].std())
print("Summary statistics:")
print(df.describe())
```

```python
print("Task-C\n")
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]
df["Percentage"] = df["TotalMarks"] / 100 * 100
print("Updated DataFrame:")
print(df)
```

```python
print("Task-D\n")
print("Students with attendance > 90%:\n",df[df["Attendance"] > 90])
print("\nStudents with total marks > 90:\n",df[df["TotalMarks"] > 90])
print("\nFemale students:\n",df[df["Gender"] == "F"])
print("\nMale students:\n",df[df["Gender"] == "M"])
print("\nStudents with external marks < 50:\n",df[df["ExternalMarks"] < 50])
```

```python
print("Task-E\n")
print("Highest Total Marks:\n",df.loc[df["TotalMarks"].idxmax()])
print("\nLowest Total Marks:\n",df.loc[df["TotalMarks"].idxmin()])
print("\nTop 3 performers:\n",df.sort_values(by="TotalMarks", ascending=False).head(3))
print("\nBottom 3 performers:\n",df.sort_values(by="TotalMarks", ascending=True).head(3))
class_avg = df["TotalMarks"].mean()
print("\nNo. of Students who scored above class average: ",df[df["TotalMarks"] > class_avg].shape[0])
```

```python
print("Task-F\n")
print("Average TotalMarks of male students:", df[df["Gender"] == "M"]["TotalMarks"].mean())
print("\nAverage TotalMarks of female students:", df[df["Gender"] == "F"]["TotalMarks"].mean())
print("\nAverage attendance of male students:", df[df["Gender"] == "M"]["Attendance"].mean())
print("\nAverage attendance of female students:", df[df["Gender"] == "F"]["Attendance"].mean())
```

```python
import matplotlib.pyplot as plt

print("Task-G\n")

plt.figure(figsize=(8, 5))
plt.bar(df["Name"], df["TotalMarks"], color="green", edgecolor="black")
plt.title("Total Marks of All Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(df["Attendance"], bins=5, color="orange", edgecolor="black")
plt.title("Histogram of Attendance")
plt.xlabel("Attendance")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 6))
gender_counts = df["Gender"].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", colors=["skyblue", "pink"],wedgeprops={"edgecolor": "black", "linewidth": 1})
plt.title("Gender Distribution")
plt.show()

plt.figure(figsize=(8, 5))
plt.scatter(df["Attendance"], df["TotalMarks"], color="red")
plt.title("Attendance vs Total Marks")
plt.xlabel("Attendance")
plt.ylabel("Total Marks")
plt.tight_layout()
plt.show()
```
