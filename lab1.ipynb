```python
import pandas as pd 

data = {
    'Rollno': [101, 102, 103, 104, 105],
    'Name': ['Ravi', 'Priya', 'Kiran', 'Anu', 'Sita'],
    'Marks': [85, 92, 78, 88, 95]}
df = pd.DataFrame(data)
display(df)
```

```python
print("The first 3 records: ")
display(df.head(3))
```

```python
print("Average of Marks:", df["Marks"].mean())
print("Highest marks: ",df["Marks"].max())
print("Lowest marks:", df["Marks"].min())
```

```python
def grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    else:
        return "C"
df["Grade"] = df["Marks"].apply(grade)
print("DataFrame with Grade:\n", df)
```

```python
import pandas as pd 

data2 = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [50000, 500, 1200, 10000],
    'Quantity': [10, 50, 30, 15]}
df = pd.DataFrame(data2)
display(df)
```

```python
df['TotalValue'] = df['Price'] * df['Quantity']
grand_total = df['TotalValue'].sum()

print(df)

print("\nProduct with highest total value:",df['TotalValue'].max())
print("\nOverall inventory Value of Products:", grand_total)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] = iris.target
df["Species"] = df["Species"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
print("\nIris DataFrame:")
print(df)

print("\nFirst Five Records:")
print(df.head())

print("\nNumber of Rows and Columns:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())

print("\nSamples in Each Class:")
print(df["Species"].value_counts())
```
