```python
import pandas as pd
df = pd.read_csv(r"C:\Users\CSE Lab\OneDrive - Amrita Vishwa Vidyapeetham\Desktop\comp_database.csv")
print(df)
```

```python
target = "class:buys_comp"

print("Prior Probabilities:")
priors = df[target].value_counts(normalize=True)
print(priors)
print()

print("Conditional Probabilities P(feature=value | class):")

features = ["Age", "Income", "Student", "Credit_rating"]

for feature in features:
    print(f"Feature: {feature}")
    probs = pd.crosstab(df[feature], df[target], normalize='columns')
    print(probs)
    print()
print("Conditional Probabilities with Laplace Smoothing:")

for feature in features:
    print(f"Feature: {feature}")
    values = df[feature].unique()
    classes = df[target].unique()
    
    result = pd.DataFrame(index=values, columns=classes)
    
    for c in classes:
        class_subset = df[df[target] == c]
        total_count = len(class_subset)
        num_values = len(values)
        
        for v in values:
            count = len(class_subset[class_subset[feature] == v])
            smoothed_prob = (count + 1) / (total_count + num_values)
            result.loc[v, c] = smoothed_prob
    
    print(result)
    print()
```
